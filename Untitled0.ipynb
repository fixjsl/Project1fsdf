```python
import torch
import numpy as np
import torch.nn as nn
import torch.optim as optim
import torch.nn.init as init
import matplotlib.pyplot as plt

```

```python
num_data = 1000
num_epoch = 500

noise = init.normal_(torch.FloatTensor(num_data,1),std = 0.2)

x = init.uniform_(torch.Tensor(num_data,1),-10,10)

y = 2*x+3

y_noise = y+noise

plt.figure(figsize=(10,10))

plt.scatter(x.numpy(),y_noise.numpy(),s=7,c='red')

plt.axis([-12,12,-25,25])

plt.show()
```

```python
model = nn.Linear(1,1)
loss_func = nn.L1Loss()

opt = optim.SGD(model.parameters(),lr=0.01)

loss_arr = []

label = y_noise
```

```python
for i in range(num_epoch):
    opt.zero_grad()

    output = model(x)

    loss = loss_func(output,label)

    loss.backward()

    opt.step()

    if i%10 ==0:
      plt.scatter(x.detach().numpy(),output.detach().numpy())
      plt.axis([-10,10,-30,30])
      plt.show()
      print(loss.data)
    loss_arr.append(loss.detach().numpy())
```

```python
plt.figure(figsize=(15,15))
plt.scatter(x.numpy(),y_noise.numpy(),s=5,c='gray')
plt.scatter(x.detach().numpy(),output.detach().numpy(),s=5,c='red')
plt.axis([-15,15,-35,35])
plt.show()
```

```python
plt.plot(loss_arr)
plt.show()
```
